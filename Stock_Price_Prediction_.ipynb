- Normalizing the data in order to Normalize the  higher range features for eg: Volume 

```python
def normalize_data(df):
    min_max_scaler = preprocessing.MinMaxScaler()
    df['Open'] = min_max_scaler.fit_transform(df.Open.values.reshape(-1,1))
    df['High'] = min_max_scaler.fit_transform(df.High.values.reshape(-1,1))
    df['Low'] = min_max_scaler.fit_transform(df.Low.values.reshape(-1,1))
    df['Volume'] = min_max_scaler.fit_transform(df.Volume.values.reshape(-1,1))
    df['Adj Close'] = min_max_scaler.fit_transform(df['Adj Close'].values.reshape(-1,1))
    df['Close'] = min_max_scaler.fit_transform(df['Close'].values.reshape(-1,1))
    return df
df = normalize_data(stockdf)
df.head()
```

# Data Preparation for Recurrent Neural Networks 

```python
# given we have values of closing price for past 5 days, we want to predict the price of next day

"This helper function is the real meat of timeseries forcasting it transform input data to the required input matrices and output vector"
def load_data(stock, seq_len):
    amount_of_features = len(stock.columns) # 5
    data = stock.to_numpy() 
    sequence_length = seq_len + 1 # index starting from 0
    result = []
    
    for index in range(len(data) - sequence_length): # maxmimum date = lastest date - sequence length
        result.append(data[index: index + sequence_length]) # index : index + 22days
    
    result = np.array(result)
    row = round(0.9 * result.shape[0]) # 90% split
    train = result[:int(row), :] # 90% date, all features 
    
    x_train = train[:, :-1] 
    y_train = train[:, -1][:,-1]
    
    x_test = result[int(row):, :-1] 
    y_test = result[int(row):, -1][:,-1]

    x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], amount_of_features))
    x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], amount_of_features))  

    return [x_train, y_train, x_test, y_test]
```

```python
# We are taking five day window to predict the output
window = 5
X_train, y_train, X_test, y_test = load_data(df, window)
```

# Helper Functions 

```python
"Below are few helper function which helps in  data transformation and normalization"

def do_difference(data, time_diff=1):
    df = []
    for row in range(time_diff, len(data)):
        df.append(data[row] - data[row - time_diff])
    return np.array(df)

def invert_difference(differenced_data, y_values):
    undiff = []
    index_to_start = len(differenced_data) + 1

    for i in range(len(differenced_data)):
        undiff.append(differenced_data[i] + y_values[-(index_to_start - i)])
    return undiff
    

def do_scaling(data):

    scaler = MinMaxScaler()
    scaler = scaler.fit(data.reshape(-1,1))
    scaledf = scaler.transform(data.reshape(-1,1))
    scaledf = scaledf.reshape(scaledf.shape[0])
    return scaler, scaledf

def invert_scaling(scaler, data):
    return scaler.inverse_transform(data).reshape(data.shape[0])

def plot_model_results(model):
    loss = model.history["loss"]
    plt.plot(loss)
    plt.ylabel('loss')
    plt.xlabel('epochs')
    plt.show()
```

```python
diff_vol = do_difference(stockdf["Volume"])
```

```python
scaler_diff_vol, scaled_diff_vol = do_scaling(diff_vol)
```

# LSTM

```python
def lstm_model(trainx, trainy, neurons, lookback, features, epochs,batch_size=1, activation_func = "tanh"):
    X, y = trainx, trainy
    model = Sequential()
    # model.add(LSTM(4, activation='relu', input_shape=(n_steps, n_features)))
    model.add(LSTM(neurons, batch_input_shape=(batch_size,lookback, features),dropout=0.1, stateful=True))
    model.add(Dense(1)) 
    model.compile(optimizer='adam', loss='mean_squared_error')
    # X = X.reshape((X.shape[0], X.shape[1], n_features))
    X = X.reshape((X.shape[0], X.shape[1], features)) 
    history = model.fit(X, y, epochs=epochs, batch_size=batch_size, shuffle=False)
    return history, model
```

```python
neurons=50
lookback=5
features=6
epochs=30
history1, model1 = lstm_model(X_train,y_train,neurons,lookback,features,epochs)
```

# GRU

```python
def GRU_model(trainx, trainy, input_neurons, hidden_neurons, lookback, features, epochs,batch_size=128, dropout=0.5):
    X, y = trainx, trainy
    model = Sequential()
    model.add(GRU (input_neurons, dropout=0.1, return_sequences = True, input_shape=(lookback, features)))
    model.add(GRU(hidden_neurons, dropout=0.1))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    X = X.reshape((X.shape[0], X.shape[1], features))
    history = model.fit(X, y, epochs=epochs, batch_size=batch_size, shuffle=False)
    return history, model
```

```python
history2 , model2 = GRU_model(X_train, y_train, 64,20,5,6,30,)
```

# Bidirectional LSTM

```python
def bidirectional_model(trainx, trainy, input_neurons, hidden_neurons, lookback, features, epochs,batch_size=128, dropout=0.5):
    X, y = trainx, trainy
    model = Sequential()
    model.add(Bidirectional(LSTM(input_neurons,return_sequences=True, dropout=0.1, input_shape=(lookback, features))))
    model.add(Bidirectional(LSTM(hidden_neurons, dropout=0.1)))
    model.add(Dense(1))  
    model.compile(optimizer='adam', loss='mean_squared_error')
    X = X.reshape((X.shape[0], X.shape[1], features))
    history = model.fit(X, y, epochs=epochs, batch_size=batch_size, shuffle=False)
    return history, model
```

```python
history3, model3 = bidirectional_model(X_train, y_train, 64,20,5,6,30,)
```

# Models Evaluation on Test data 

```python
"These are the helper function to evalute the testing performance of RNN models"

def predict_data(model, testx, features=1):
    predictx = testx.reshape(testx.shape[0], testx.shape[1], features)
    y_pred = model.predict(predictx,batch_size=1)
    return y_pred

def plot_predicted(testy, y_pred, original_y, model_name, scaler):
    
    y_pred_unscaled = scaler.inverse_transform(y_pred)
    testy_unscaled = scaler.inverse_transform(testy.reshape(testy.shape[0],1))
    testy_undiff = invert_difference(testy_unscaled,original_y)
    pred_y_undiff = invert_difference(y_pred_unscaled, original_y)
    plt.plot(testy_undiff, label="Actual")
    plt.plot(pred_y_undiff, label="Predicted")
    plt.legend(loc="best")
    plt.title("Model Name = " + str(model_name))

def evaluate_results(testy, y_pred, original_y, model_name, scaler):
    errors = y_pred - testy
    rmse = np.sqrt(np.square(errors).mean())
    mae = np.mean(np.abs(errors))
    print('Root Mean Square Error = ' + str(rmse))
    print("mean Absolute Error = " + str(mae))
    plot_predicted(testy, y_pred, original_y, model_name, scaler)

```

```python
lstm_pred = predict_data(model1,X_test,6)
bidirec_pred= predict_data(model2,X_test,6)
gru_pred = predict_data(model3,X_test,6)
```

```python
fig, axs = plt.subplots()
fig.set_figheight(10)
fig.set_figwidth(15)
evaluate_results(y_test,lstm_pred,stockdf["Volume"], "LSTM",scaler_diff_vol)
plt.show()
fig, axs = plt.subplots()
fig.set_figheight(10)
fig.set_figwidth(15)
evaluate_results(y_test,bidirec_pred,stockdf["Volume"], "Bi Directed LSTM",scaler_diff_vol)
plt.show()

fig, axs = plt.subplots()
fig.set_figheight(10)
fig.set_figwidth(15)
evaluate_results(y_test,gru_pred,stockdf["Volume"], "GRU",scaler_diff_vol)
```
